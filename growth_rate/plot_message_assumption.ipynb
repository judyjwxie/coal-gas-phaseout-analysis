```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
growth_up = pd.read_excel("message_assumption.xlsx",sheet_name="growth_new_capacity_up")
growth_lo = pd.read_excel("message_assumption.xlsx",sheet_name="growth_new_capacity_lo")
growth_up
```

```python
growth_up.technology.unique()
```

```python
fig, axes = plt.subplots(figsize=(3,3))
sns.lineplot(x='year_vtg', y='value', data=growth_up.query("technology=='gas_t_d'"),style="node_loc",hue="technology",ax=axes)
axes.set_ylabel("growth_new_capacity_up")
axes.legend(loc='center left', bbox_to_anchor=(1, 0.5),frameon=False)
```

```python
fig, axes = plt.subplots(figsize=(3,3))
sns.scatterplot(x='year_vtg', y='value', data=growth_lo,style="node_loc",hue="technology",ax=axes)
axes.set_ylabel("growth_new_capacity_lo")
axes.legend(loc='center left', bbox_to_anchor=(1, 0.5),frameon=False)
```
