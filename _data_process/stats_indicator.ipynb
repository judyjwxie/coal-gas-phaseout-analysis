```python
import pandas as pd
import numpy as np
```

```python
coal_data = pd.read_csv("_all_temporal_coal_indicators.csv")
coal_data
```

```python
gas_data = pd.read_csv("_all_temporal_gas_indicators.csv")
```

```python
print(len(coal_data["Country"].unique()))
print(coal_data["Year"].min())
print(coal_data["Year"].max())
```

```python
print(len(gas_data["Country Code"].unique()))
print(gas_data["year"].min())
print(gas_data["year"].max())
```

```python
for ind in ["Coal_Share_Ret","Coal_Share_Tran","Coal_Share_PhasedOut","Gas_Share_Ret",
            "Coal_duration_year","Gas_duration_year","Coal_logCurrent_MW","Gas_logCurrent_MW"]:
    if "Gas" in ind:
        new_df = pd.DataFrame(data={"Mean":gas_data[ind].mean(), "STD":gas_data[ind].std(),
                               "Min":gas_data[ind].min(),"Max":gas_data[ind].max(),
                                    "N":len(gas_data[ind])},index=[ind])
    else:
        new_df = pd.DataFrame(data={"Mean":coal_data[ind].mean(), "STD":coal_data[ind].std(),
                               "Min":coal_data[ind].min(),"Max":coal_data[ind].max(),
                                    "N":len(coal_data[ind])},index=[ind])
    if ind == "Coal_Share_Ret":
        stats = new_df
    else:
        stats = pd.concat([stats,new_df])
stats = round(stats,2)
stats
```
