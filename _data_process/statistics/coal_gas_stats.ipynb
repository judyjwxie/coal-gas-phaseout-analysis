```python
import pandas as pd
import numpy as np
```

```python
gas_cap_data = pd.read_csv("../_all_temporal_gas_indicators.csv")
gas_var_list = ["Gas_Current_MW","Gas_Share_Ret","Gas_Share_PhasedOut"]
gas_stats = pd.DataFrame(data={"Variable":gas_var_list,"Mean":np.nan,
                  "STD":np.nan, "Min":np.nan, "Max":np.nan, "N":np.nan
                  }).set_index("Variable")
gas_stats["N"] = len(gas_cap_data)
gas_stats["Year Range"] = str(gas_cap_data["year"].min())+"-"+str(gas_cap_data["year"].max())
gas_stats["Number of Countries"] = len(gas_cap_data["Country Code"].unique())
for i in gas_var_list:
    gas_stats.loc[i,"Mean"] = round(gas_cap_data[i].mean(),2)
    gas_stats.loc[i,"STD"] = round(gas_cap_data[i].std(),2)
    gas_stats.loc[i,"Min"] = round(gas_cap_data[i].min(),2)
    gas_stats.loc[i,"Max"] = round(gas_cap_data[i].max(),)
gas_stats
```

```python
coal_cap_data = pd.read_csv("../_all_temporal_coal_indicators.csv")
coal_var_list = ["Coal_Current_MW","Coal_Share_Ret","Coal_Share_Tran","Coal_Share_PhasedOut"]
coal_stats = pd.DataFrame(data={"Variable":coal_var_list,"Mean":np.nan,
                  "STD":np.nan, "Min":np.nan, "Max":np.nan
                  }).set_index("Variable")
coal_stats["N"] = len(coal_cap_data)
coal_stats["Year Range"] = str(coal_cap_data["Year"].min())+"-"+str(coal_cap_data["Year"].max())
coal_stats["Number of Countries"] = len(coal_cap_data["Country"].unique())
for i in coal_var_list:
    coal_stats.loc[i,"Mean"] = round(coal_cap_data[i].mean(),2)
    coal_stats.loc[i,"STD"] = round(coal_cap_data[i].std(),2)
    coal_stats.loc[i,"Min"] = round(coal_cap_data[i].min(),2)
    coal_stats.loc[i,"Max"] = round(coal_cap_data[i].max(),)
coal_stats
```

```python
coalgas_data = pd.read_csv("../../../analysis/temporal_country_models/_all_temporal_coalgas_var.csv")
coalgas_var_list = ['Share_VRE_Gen', 'Share_Gas_Gen', 'Share_Coal_Gen', 'Coal_Gen_Decline_Perc', 'Gas_Gen_Decline_Perc']
coalgas_stats = pd.DataFrame(data={"Variable":coalgas_var_list,"Mean":np.nan,
                  "STD":np.nan, "Min":np.nan, "Max":np.nan
                  }).set_index("Variable")
coalgas_stats["N"] = len(coalgas_data)
coalgas_stats["Year Range"] = str(int(coalgas_data["year"].min()))+"-"+str(int(coalgas_data["year"].max()))
coalgas_stats["Number of Countries"] = len(coalgas_data["Country Code"].unique())
for i in coalgas_var_list:
    coalgas_stats.loc[i,"Mean"] = round(coalgas_data[i].mean(),2)
    coalgas_stats.loc[i,"STD"] = round(coalgas_data[i].std(),2)
    coalgas_stats.loc[i,"Min"] = round(coalgas_data[i].min(),2)
    coalgas_stats.loc[i,"Max"] = round(coalgas_data[i].max(),)
coalgas_stats
```

```python
all_data = pd.concat([gas_stats, coal_stats, coalgas_stats])
all_data.to_csv("coal_gas_stats.csv")
```
