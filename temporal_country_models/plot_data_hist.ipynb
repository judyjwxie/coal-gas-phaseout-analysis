```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

# Histogram of variables

```python
full_data = pd.read_csv("coal_original_data.csv")
full_data = full_data.drop(columns="Unnamed: 0")
full_data["type"] = "original"
listwise_delete = pd.read_csv("coal_original_data_listwisedel.csv")
listwise_delete = listwise_delete.drop(columns="Unnamed: 0")
listwise_delete["type"] = "list-wise deletion"
combined_data = pd.concat([full_data, listwise_delete])
combined_data = combined_data.reset_index().drop(columns="index")
combined_data
```

```python
col_names = full_data.columns[2:-1]
fig, axes = plt.subplots(nrows=10, ncols=5,figsize=(10,20))
plt.subplots_adjust(wspace=0.4,hspace=0.4)
for i, ax in enumerate(fig.axes):
    if i >=49:
        continue
    col = col_names[i]
    sns.kdeplot(data = combined_data, x=col, hue = "type" , ax=ax, legend=None)
    ax.set_title(col)
    ax.set_xlabel(None)
    ax.set_ylabel(None)
```
